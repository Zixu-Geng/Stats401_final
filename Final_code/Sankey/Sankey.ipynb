```python
clean_data = clean[['TEWHERE', 'TUACTDUR', 'TRCODE']]
```

```python


def label(df):
    if df["TUACTDUR"] < 60:
        return 60
    elif df["TUACTDUR"] < 120:
        return 120
    elif df["TUACTDUR"] < 180:
        return 180
    elif df["TUACTDUR"] < 240:
        return 240
    else :
        return 1000
clean_data['TUACTDUR']=clean_data.apply(label,axis=1)

link_data = clean_data
```
